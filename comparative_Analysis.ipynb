```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Data
models = ['ConvNeXt-V2', 'EfficientNet-B0', 'Hybrid (EfficientNet-B0 + ConvNeXt-V2)', 'AgroSpectraNet']
train_acc = [99.66, 99.88, 99.77, 99.91]
val_acc = [95.59, 96.75, 96.96, 97.11]
test_acc = [94.71, 96.04, 95.36, 96.26]

x = np.arange(len(models))  # the label locations
width = 0.25  # the width of the bars

# Plot
fig, ax = plt.subplots(figsize=(12,6))

# Using distinct, visually-friendly colors
rects1 = ax.bar(x - width, train_acc, width, label='Train Accuracy', color='#1f77b4')   # blue
rects2 = ax.bar(x, val_acc, width, label='Validation Accuracy', color='#ff7f0e')       # orange
rects3 = ax.bar(x + width, test_acc, width, label='Test Accuracy', color='#2ca02c')     # green

# Labels and formatting
ax.set_ylabel('Accuracy (%)')
ax.set_xlabel('Models')
ax.set_title('Comparison of Training, Validation, and Test Accuracies')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=15, ha='right')
ax.set_ylim(90, 102)
ax.legend()

# Adding value labels above bars
def add_labels(rects, offset=3):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, offset),
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9)

add_labels(rects1)
add_labels(rects2)
add_labels(rects3)

plt.tight_layout()

# Save the figure
save_path = r"C:\Users\ZA SEYAM\Documents\AgroSpectraNet_accuracy_comparison.png"  # change path if needed
os.makedirs(os.path.dirname(save_path), exist_ok=True)
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"Figure saved at: {save_path}")

# Show the plot
plt.show()
```
